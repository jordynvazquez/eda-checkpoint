## Visualize

Create any visualizations for your EDA here. Make note in the form of code comments of what your thought process is for your visualizations.

```python
# Showing the distribution of the amount of shows released each year. 
plt.style.use('dark_background')
fig1 = tv_df['Year Released'].value_counts()
plt.figure(figsize = (14,15))
sb.barplot(x = fig1, y = fig1.index, palette = 'rainbow_r', orient = 'h')
plt.title('Streaming Shows Over the Years', fontsize = 20)
plt.xlabel('Number of Shows', fontsize = 15)
plt.ylabel('Years', fontsize = 15)
plt.yticks(fontsize = 7)
plt.xticks(fontsize = 10, rotation = 45)
plt.show()
```

```python
# Showing what the distribution for content rating is across all platforms
plt.style.use('dark_background')
fig2 = tv_df['Content Rating'].value_counts()
plt.figure(figsize = (14,15))
sb.barplot(x = fig2.index, y = fig2, palette = 'rainbow_r')
plt.title('Content Rating Across All Platforms', fontsize = 20)
plt.xlabel('Content Rating', fontsize = 15)
plt.ylabel('Number of Shows', fontsize = 15)
plt.yticks(fontsize = 7)
plt.xticks(fontsize = 10, rotation = 45)
plt.show()
```

```python
# Cannot visualize the Genre and Streaming Platform columns until after I clean the data since a lot of them have multiple values. 
# I made a word could to visualize the top listed Genres
from wordcloud import WordCloud, STOPWORDS

fig3 = " ".join(review for review in tv_df.Genre.astype(str))
stopwords = set(STOPWORDS)
wordcloud = WordCloud(background_color = 'white', max_words = 200, stopwords = stopwords)

wordcloud.generate(fig3)
plt.imshow(wordcloud, interpolation = 'bilinear')
plt.axis('off')
plt.show()
```
